# Sub Assignment 2: Exploring And Visualizing Indian States Data

```python
# 1. What is the population of each state in India?

state_population = df.groupby('state_name')['Population'].sum().reset_index()

# Sort the data by population in descending order for better visualization
state_population_sorted = state_population.sort_values(by='Population', ascending=False)

# Create a bar chart illustrating the population distribution
plt.figure(figsize=(12, 6))
plt.bar(state_population_sorted['state_name'], state_population_sorted['Population'], color='skyblue')
plt.xticks(rotation=90)
plt.title("Population of Each State in India")
plt.xlabel("state_name")
plt.ylabel("Population")
plt.show()
```

```python
# 2. What is the literacy rate in different states?

# age_literacy_rate = df.groupby('Age_Group_0_29')['Literate'].mean().reset_index()

# # Create a line chart
# plt.figure(figsize=(10, 6))
# plt.plot(age_literacy_rate['Age_Group_0_29'], age_literacy_rate['Literate'], marker='o', linestyle='-', color='purple')
# plt.title('Literacy Rate by Age Group in India')
# plt.xlabel('Age Group')
# plt.ylabel('Average Literacy Rate')
# plt.xticks(rotation=45)
# plt.show()


# Group the data by state and calculate the average literacy rate
state_literacy_rate = df.groupby('state_name')['Literate'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Literate', y='state_name', data=state_literacy_rate, palette='viridis')
plt.title("Literacy Rate in Different States in India")
plt.xlabel("Literate")
plt.ylabel("State & District")
plt.show()
```

```python
# 3. What is the gender distribution across different states in India?

gender_distribution = df.groupby('state_name')[['Male', 'Female']].sum()

# Create a stacked bar chart
gender_distribution.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Gender Distribution by State in India')
plt.xlabel('state_name')
plt.ylabel('Count')
plt.show()
```

```python
# 4. How does the number of educated individuals vary across states?

education_distribution = df.groupby('state_name')[['Primary_Education', 'Secondary_Education', 'Higher_Education']].sum()

# Create a stacked bar chart
education_distribution.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Education Level Distribution by State in India')
plt.xlabel('state_name')
plt.ylabel('Count')
plt.show()
```

```python
# 5. What is the distribution of the population in different age groups?

plt.figure(figsize=(10, 6))
plt.hist(df['Age_Group_0_29'], bins=10, edgecolor='k', alpha=0.7)
plt.title('Age Distribution of Population in India')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
#  6. How does the literacy rate vary with age?

age_literacy_rate = df.groupby('Age_Group_0_29')['Literate'].mean().reset_index()

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(age_literacy_rate['Age_Group_0_29'], age_literacy_rate['Literate'], marker='*', linestyle='-', color='purple')
plt.title('Literacy Rate by Age Group in India')
plt.xlabel('Age Group')
plt.ylabel('Average Literacy Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. What is the distribution of different educational levels in the population by state?

# Group the data by state and calculate the count of individuals with different levels of education
education_distribution = df.groupby('state_name')[['Primary_Education', 'Secondary_Education', 'Higher_Education']].sum()

# Create a stacked bar chart
education_distribution.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Education Level Distribution by State in India')
plt.xlabel('state_name')
plt.ylabel('Count')
plt.show()
```

```python
# 8. What is the distribution of Scheduled Castes (SC) by gender in each state?

# Group the data by state and calculate the count of SC individuals by gender
sc_gender_distribution = df.groupby('state_name')[['Male_SC', 'Female_SC']].sum()

# Create a stacked bar chart
sc_gender_distribution.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Distribution of Scheduled Castes (SC) by Gender in Each State in India')
plt.xlabel('state_name')
plt.ylabel('Count')
plt.show()
```

```python
# 9. What is the distribution of education levels among different states in India?

education_distribution = df.groupby('state_name')[['Illiterate_Education', 'Literate_Education', 'Graduate_Education', 'Other_Education']].sum()

# Create a stacked bar chart
plt.figure(figsize=(12, 9))
education_distribution.plot(kind='bar', stacked=True)
plt.title('Distribution of Education Levels in Different States of India')
plt.xlabel('State & District')
plt.ylabel('Population')
plt.legend(title='Education Level')
plt.xticks(rotation=90)
plt.show()
```

```python
# 10. What is the distribution of the population in different age groups in India?

# Filter the data for the top 20 states by population
top_20_states = df.nlargest(20, 'Population')

# Group the filtered data by state and age group, then calculate the average literacy rate
state_age_literacy_rate = top_20_states.groupby(['state_name', 'Age_Group_0_29'])['Literate'].mean().reset_index()

# Create subplots for each state's age group distribution
plt.figure(figsize=(15, 15))
counter = 1  # Counter for subplot position

for state in top_20_states['state_name']:
    state_data = state_age_literacy_rate[state_age_literacy_rate['state_name'] == state]
    plt.subplot(5, 4, counter)
    plt.pie(state_data['Literate'], labels=state_data['Age_Group_0_29'], autopct='%1.1f%%', startangle=140)
    plt.title(f'{state}\nAge Group Distribution')
    plt.axis('equal')
    
    counter += 1

# Add space between subplots
plt.subplots_adjust(left=0.2, right=0.9, bottom=0.2, top=0.9, wspace=0.8, hspace=0.8)

# Display the image directly in the Colab notebook
plt.show()
```
